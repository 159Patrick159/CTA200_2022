# Q2 Lorenz Equation

$$
\dot{X} = -\sigma(X - Y)\\
\dot{Y} = rX - Y - XZ\\
\dot{Z} = -bZ + XY
$$

```python
# Define the system of equation
def lorenz_equations(t,y,sigma,r,b):
    X,Y,Z = y 
    dotX = -sigma*(X-Y)
    dotY = r*X - Y - X*Z
    dotZ = -b*Z + X*Y
    return([dotX,dotY,dotZ])
```

```python
from scipy.integrate import ode

# Set intials conditions
W0 = [0,1,0]
t0 = 0

# Set parameteres
sig, r, b = 10, 28, 8/3

# Set integration bounds
t_end = 60
dt = 0.01

# Array to store the results
times = [t0]
res = [W0]

solver = ode(lorenz_equations).set_initial_value(W0,t0).set_f_params(sig,r,b)

# Solve the equations
while solver.successful() and solver.t < t_end:
    times.append(solver.t + dt)
    res.append(solver.integrate(solver.t + dt))

# Store the solutions in arrays
X_sol, Y_sol, Z_sol = np.array(res).T
times = np.array(times)
```

```python
# Slice solutions to specified bounds
N = times/dt

figure, (a0,a1,a2) = plt.subplots(figsize=(6.5,4),nrows=3,dpi=120)
a0.set_title("Solution of Y for given iterations")
a0.plot(N[:1000],Y_sol[:1000],c="k")
a1.plot(N[1000:2000],Y_sol[1000:2000],c='k')
a2.plot(N[2000:3000],Y_sol[2000:3000],c='k')
a2.set_xlabel(r"Iteration Number $(t/\Delta t)$")
plt.tight_layout()
plt.savefig("Y_Solutions.png")
plt.show()
```
